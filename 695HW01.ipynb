```python
import numpy as np 
import matplotlib.pyplot as plt

plt.style.use('seaborn-whitegrid')
```

# 1: Noisy and Polynomial Plot

```python
#original 100 
noise_scale = 100 

number_of_samples = 50 

x = 25*(np.random.rand(number_of_samples, 1) - 0.8)
y = 7*x + 15*x**2 + 1*x**3 + noise_scale*np.random.randn(number_of_samples, 1)

y2 = 7*x + 15*x**2 + 1*x**3



xf = x.flatten()
yf = y.flatten()


def polyplotter(m, x, y):
    params = np.polyfit(x,y, m)
    
    total = 0
    i = 0
    while m >= 0:
        total += params[m]*x**(i)
        m -= 1
        i += 1
        
    return total
    
mse_array = np.array([])    

for i in range(1,8):
    plt.plot(x,y,'ro',label = 'w/ noise')
    plt.plot(x,y2,'bx',label = 'real')

    plotter = polyplotter(i, xf, yf)
    plt.plot(xf,plotter,'gx', label = 'order {} fit'.format(i))
    plt.legend()
    #plt.title("Order 2")
    plt.show() 
    

    plotter = np.array(plotter)
    yf = np.array(yf)
    
    
    diff = yf - plotter
    ans = 1/len(plotter)*np.dot(diff, diff)
    
    #print(ans)
    
    mse_array = np.append(mse_array,ans)
#print(mse_array)


x1 = [1,2,3,4,5,6,7,8]
y1 = mse_array

#UNCOMMENT TO GET MSE VS M 
#plt.scatter(x1,y1)
#plt.plot(x1,y1)

#plt.show()

```

# 3: Change Noise Scale

```python
#original 100 
noise_scale = np.array([150, 200, 400, 600, 1000]) 

number_of_samples = 50 
for j in range(len(noise_scale)):
    x = 25*(np.random.rand(number_of_samples, 1) - 0.8)
    y = 7*x + 15*x**2 + 1*x**3 + noise_scale[j]*np.random.randn(number_of_samples, 1)

    y2 = 7*x + 15*x**2 + 1*x**3



    xf = x.flatten()
    yf = y.flatten()


    def polyplotter(m, x, y):
        params = np.polyfit(x,y, m)

        total = 0
        i = 0
        while m >= 0:
            total += params[m]*x**(i)
            m -= 1
            i += 1

        return total

    mse_array = np.array([])    

    for i in range(3,4):
        plt.plot(x,y,'ro',label = 'noise = {}'.format(noise_scale[j]))
        #plt.plot(x,y2,'bx',label = 'real')

        plotter = polyplotter(i, xf, yf)
        plt.plot(xf,plotter,'gx', label = 'order {} fit'.format(i))
        plt.legend()
        plt.show() 


        plotter = np.array(plotter)
        yf = np.array(yf)


        diff = yf - plotter
        ans = 1/len(plotter)*np.dot(diff, diff)

        #print(ans)

        mse_array = np.append(mse_array,ans)
    #print(mse_array)
```

As the noise_scale variable increases, the data points become more spread out and less related to each other, leading to the fitted polynomial that fits the data less accurately. 

# 4: Change Number of Samples

```python
#original 100 
noise_scale = 100
#original = 50
number_of_samples = np.array([ 40, 30, 20, 10])

for j in range(len(number_of_samples)):
    x = 25*(np.random.rand(number_of_samples[j], 1) - 0.8)
    y = 7*x + 15*x**2 + 1*x**3 + noise_scale*np.random.randn(number_of_samples[j], 1)

    y2 = 7*x + 15*x**2 + 1*x**3



    xf = x.flatten()
    yf = y.flatten()


    def polyplotter(m, x, y):
        params = np.polyfit(x,y, m)

        total = 0
        i = 0
        while m >= 0:
            total += params[m]*x**(i)
            m -= 1
            i += 1

        return total

    mse_array = np.array([])    

    for i in range(3,4):
        plt.plot(x,y,'ro',label = 'samples = {}'.format(number_of_samples[j]))
        #plt.plot(x,y2,'bx',label = 'real')

        plotter = polyplotter(i, xf, yf)
        plt.plot(xf,plotter,'gx', label = 'order {} fit'.format(i))
        plt.legend()
        plt.show() 


        plotter = np.array(plotter)
        yf = np.array(yf)


        diff = yf - plotter
        ans = 1/len(plotter)*np.dot(diff, diff)

        #print(ans)

        mse_array = np.append(mse_array,ans)
    #print(mse_array)
```

As the number of samples decreases, the fitted polynomial curve becomes less accurate. This is because as the number of samples decreases then the polynomial has less data to work with and create an accurate curve. We can see that when there are only ten samples, the data points are few and far between and the polynomial does not create a clear curve.
